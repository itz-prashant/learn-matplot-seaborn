```python
plt.plot(x, y)
plt.title("Hello")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
```

```python
plt.subplot(2,2,1)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
```
